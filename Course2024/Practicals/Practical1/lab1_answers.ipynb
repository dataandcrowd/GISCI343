### Step 5: Creating and Modifying Lists

```python
cities = ["Auckland", "Wellington", "Christchurch"]
print(cities)
cities.append("Hamilton")
print(cities)
cities[1] = "Dunedin"
print(cities)
```

### Step 6: For Loops

```python
for city in cities:
    print(city)
```


* Double loop

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]
pairwise_sum = []

for i in list1:
    for j in list2:
        pairwise_sum.append(i + j)

print(pairwise_sum)

```

### Step 7: Combining Loops and Conditionals
* Use loops and conditionals together

```python
temperatures = [20, 25, 30, 35]
for temp in temperatures:
    if temp > 30:
        print(temp, "is hot")
    else:
        print(temp, "is not hot")
```

## Part 3: Practical Exercises

### Exercise 1: Calculating Mean Temperature

* Calculate the mean temperature from a list

```python
temperatures = [20, 22, 24, 26, 28, 30]
mean_temp = sum(temperatures) / len(temperatures)
print("Mean Temperature:", mean_temp)
```
